```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# load apple watch data
comedy_aw = pd.read_csv('generated_data/clara/comedy_aw.csv')
documentary_aw = pd.read_csv('generated_data/clara/documentary_aw.csv')
horror_aw = pd.read_csv('generated_data/clara/horror_aw.csv')

# load tracked motion data
comedy_tm = pd.read_csv('generated_data/clara/comedy_tm.csv')
documentary_tm = pd.read_csv('generated_data/clara/documentary_tm.csv')
horror_tm = pd.read_csv('generated_data/clara/horror_tm.csv')

# load pupil size data
comedy_ps = pd.read_csv('generated_data/clara/comedy_ps.csv')
documentary_ps = pd.read_csv('generated_data/clara/documentary_ps.csv')
horror_ps = pd.read_csv('generated_data/clara/horror_ps.csv')
```

```python
def merge_dataframes(aw_df, tm_df, ps_df):
    # Convert to datetime
    aw_df['timestamp'] = pd.to_datetime(aw_df['timestamp']).dt.tz_localize(None)
    tm_df['timestamp'] = pd.to_datetime(tm_df['timestamp']).dt.tz_localize(None)
    ps_df['timestamp'] = pd.to_datetime(ps_df['timestamp']).dt.tz_localize(None)
    
    # Sort all DataFrames 
    aw = aw_df.sort_values('timestamp')
    tm = tm_df.sort_values('timestamp')
    ps = ps_df.sort_values('timestamp')
    
    # Merge on nearest time
    merged = pd.merge(aw, tm, on='timestamp', how='right')
    return pd.merge(merged, ps, on='timestamp', how='inner')
```

```python
comedy = merge_dataframes(comedy_aw, comedy_tm, comedy_ps)
documentary = merge_dataframes(documentary_aw, documentary_tm, documentary_ps)
horror = merge_dataframes(horror_aw, horror_tm, horror_ps)
```

```python
# Add genre column
comedy['genre'] = 'comedy'
documentary['genre'] = 'documentary'
horror['genre'] = 'horror'

# Concatenate all
df_all = pd.concat([comedy, documentary, horror], ignore_index=True)
```
