```python
from model_pipeline import model_pipeline
from xgboost import XGBClassifier
import pandas as pd

# Load your data
df = pd.read_csv("../../feature_engineering/data_with_new_features_v3.csv")

# Run the pipeline, this time with XGBClassifier under the 'clf' step
results_xgb = model_pipeline(
    df,
    model_factory=lambda **p: XGBClassifier(
        use_label_encoder=False,
        eval_metric='mlogloss',
        random_state=42,
        **p
    ),
    normalize='global',
    test_val_split=0.1,
    param_distributions={
        'clf__n_estimators':        (50, 1000, 50),     # number of trees
        'clf__max_depth':           (3, 15, 1),         # depth per tree
        'clf__learning_rate':       (0.01, 0.3, 0.01),   # shrinkage
        'clf__subsample':           (0.5, 1.0, 0.1),     # row subsampling
        'clf__colsample_bytree':    (0.5, 1.0, 0.1),     # feature subsampling
        'clf__gamma':               (0.0, 5.0, 0.5),     # min loss reduction to split
        'clf__reg_alpha':           (0.0, 5.0, 0.5),     # L1 term on weights
        'clf__reg_lambda':          (0.0, 5.0, 0.5),     # L2 term on weights
    },
    n_trials=1,
    search_cv=3,
    test_participant='aimen'
)

print("Aimen test macro-F1 with XGB:", results_xgb['reports']['aimen']['f1_macro'])
```
