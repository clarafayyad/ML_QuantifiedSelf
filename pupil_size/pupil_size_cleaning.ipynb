```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
pupil_size_csv = 'pupil_size_extracted_horror.csv'
start_time = datetime.strptime("2025-06-05 17:58:48", "%Y-%m-%d %H:%M:%S")
```

```python
df = pd.read_csv(pupil_size_csv)
df.head()
```

```python
# Convert timestamp (float seconds) to actual datetime
df['time'] = df['timestamp'].apply(lambda x: start_time + timedelta(seconds=x))
df.head()
```

```python
df = df.sort_values(by='time').reset_index(drop=True)
df['timestamp'] = df['time'] 
df.drop(columns=['time'], inplace=True)
```

```python
df
```

```python
df = df.set_index('timestamp')  # or 'time' if that's your column name
df
```

```python
df.to_csv('../data_collection/generated_data/pupil_size_horror.csv')
```
